```python
import pandas as pd
```

```python
df = pd.read_table("P3-06-Approved_smi_inchi.txt", header=None)
```

```python
approved_drugs = df[0]
len(approved_drugs.tolist())
```

```python
df = pd.read_table("P1-03-Drug_Synonyms.txt", header=None)
df = df[df[0].isin(approved_drugs.tolist())]
df[1] = df[1].astype(str) + "; "
```

```python
df[1] = df[1].map(str) + df[2]
```

```python
df.drop(2,1) # frist number is the column being droped and the second number indicates its a column not a row
```

```python
df2 = pd.read_table("P1-04-Drug_disease.txt")
df2
```

```python
df = pd.merge(df, df2,  how='inner', left_on=[0], right_on = ['TTDDRUGID'])
```

```python
df = df.drop(2,1)
df = df.drop('TTDDRUGID',1)
df = df.drop('LNM',1)
```

```python
df.rename(index=str, columns={1: "drug names", 0: "TTDDRUGID"})
```

```python
df.to_csv('holygrail.tsv', sep = "/t")
```
